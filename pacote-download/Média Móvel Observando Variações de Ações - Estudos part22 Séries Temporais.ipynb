# Média Móvel Observando Variações de Ações - Estudos part22 Séries Temporais

```python
from datetime import datetime
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6
```

```python
stock = pd.read_csv('all-stocks-2006-01-01-to-2018-01-01.csv', index_col='Date', parse_dates=['Date'])
```

```python
stock.head(8)
```

```python
# Filtrando os dados
g_stk = stock.query('Name == "GOOGL"')
```

```python
g_stk.head()
```

```python
g_stk['Ticks'] = range(0,len(g_stk.index.values))
```

```python
g_stk.head()
```

```python
fig = plt.figure(1)
ax1 = fig.add_subplot(111)
ax1.set_xlabel('Ticks', color='white')
ax1.set_ylabel('Preço', color='white')
ax1.set_title('Gráfico Original', fontsize=24, color='white')
ax1.plot('Ticks', 'Open', data=g_stk)
```

```python
# Gerando a média movel
g_stk['Rolling_Mean'] = g_stk['Open'].rolling(window=80).mean()
g_stk.head(8)
```

```python
# Filtrando 
filt_g = g_stk[(g_stk['Ticks'] >= 900) & (g_stk['Ticks'] <= 1100)]
filt_g.head()
```

```python
# Plotando lado a lado os dois graficos para ver o resultado
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15,5))
axes[0].plot('Ticks', 'Open', data=g_stk)
axes[0].set_title('Original', color='white')

axes[1].plot('Ticks', 'Rolling_Mean', data=g_stk)
axes[1].set_title('Média Móvel', color='white')
```
