# Componentes e Decomposição de uma Série Temporal - Estudos part16 Séries Temporais

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from pandas import Series
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15,6
from pandas.plotting import register_matplotlib_converters 
register_matplotlib_converters()
```

```python
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
# OBS: Através da decomposição podemos visualizar de forma mais clara os componetes
```

```python
# Visualizando os dados 
data.head()
```

```python
# Plotando a série sem fazer a decomposição (series original)
plt.plot(data)
plt.title('Passageiros Internacionais Transportados', color='white', fontsize=24)
plt.ylabel('Passageiros')
plt.xlabel('Data')
plt.show()
```

```python
# Para fazer a Decomposição importamos um pacote
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Fazendo a decomposição
# Armazeando dentro da variavel a decomposição sazonal com o modelo multiplicativo
result = seasonal_decompose(data, model='multiplicative')
```

```python
result.plot()
plt.show()
```

```python
# Armazeando dentro da variavel a decomposição sazonal com o modelo aditive
result2 = seasonal_decompose(data, model='aditive')
```

```python
result2.plot()
plt.show()
```
