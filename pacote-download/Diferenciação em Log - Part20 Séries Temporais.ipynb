```python
# importando os módulos
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from statsmodels.tsa.stattools import adfuller
from pandas import Series
```

```python
# Importando os dados
dateparse = lambda dates: pd.datetime.strptime(dates, '%Y-%m')
data = pd.read_csv('AirPassengers.csv', parse_dates=['Month'], index_col='Month', date_parser=dateparse)
```

```python
# Vendo algumas estatiscas usando medias moveis 
rolmean =data.rolling(window=12).mean()
rolstd = data.rolling(window=12).std()
print(rolmean)
```

```python
print(rolstd)
```

```python
# gerando o log
indexedDataset_logScale = np.log(data)
plt.plot(indexdDataset_logScale)
```

```python
# Gerando a média movel e plotando os dois juntos
movingAverenge = indexedDataset_logScale.rolling(window=12).mean()
movingSTD = indexedDataset_logScale.rolling(window=12).std()
# Plotando os dados com o log e plotar no mesmo grafico a linha do grafico da media movel
plt.plot(indexedDataset_logScale)
plt.plot(movingAverenge, color='red')
```

```python
datasetLogScaleMinusMovingAverage = indexedDataset_logScale - movingAverenge
datasetLogScaleMinusMovingAverage.head(12)
```

```python
datasetLogScaleMinusMovingAverage.dropna(inplace=True)
datasetLogScaleMinusMovingAverage.head(10)
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
def test_stationarity(timeseries):
    movingAverage = timeseries.rolling(window=12).mean()
    movingSTD = timeseries.rolling(window=12).std()
    
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage, color='red', label='Média Móvel')
    std = plt.plot(movingSTD, color='black', label='Desvio Padrão')
    plt.legend(loc='best')
    plt.title('Média Móvel E Desvio Padrão', fontsize=24, color='white')
    
    print('Resultados do teste de Dickey-Fuller:')
    dftest = adfuller(timeseries['#Passengers'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Estatística de Teste', 'Valor de P', '# de Intervalos', '# de Observações'])
    for key, value in dftest[4].items():
        dfoutput['Valor Crítico ({})'.format(key)] = value
    print(dfoutput)  
```

```python
test_stationarity(datasetLogScaleMinusMovingAverage)
```
