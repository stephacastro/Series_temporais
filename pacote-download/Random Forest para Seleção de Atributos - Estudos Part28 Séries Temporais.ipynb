# Random Forest para Seleção de Atributos - Estudos Part28 Séries Temporais
* Usando essa técnica para identificar periodos importantes

```python
import pandas as pd
from pandas import read_csv
from matplotlib import pyplot
from pandas import Series
```

```python
d = read_csv('AirPassengers.csv', header=0, parse_dates=[0], index_col=0, squeeze=True)
d
```

```python
# removendo o efeito sazonal aplicando diferenciação tornando os dados estacionarios
differenced = d.diff(12)
differenced.head(13)
```

```python
differenced.plot()
pyplot.show()
```

```python
# removendo valores nulos
differenced = differenced[12:]
```

```python
differenced.head(n=13)
```

```python
differenced.plot()
pyplot.show()
```

```python
# reorganizando os dados para prever as observações
dataframe = pd.DataFrame()
for i in range(12,0,-1):
    dataframe['t-'+str(i)] = differenced.shift(i)
    dataframe['t'] = differenced.values
print(dataframe.head(13))
dataframe = dataframe[13:]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
array = dataframe.values
X = array[:,0:-1]
y = array[:,-1]
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=1)
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
names = dataframe.columns.values[0:-1]
ticks = [i for i in range(len(names))]
pyplot.bar(ticks, model.feature_importances_)
pyplot.xticks(ticks, names)
pyplot.show()
```

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(RandomForestRegressor(n_estimators=500, random_state=1))
fit = rfe.fit(X,y)
```

```python
# crindo uma rotina de impressão para ver o resultado
names = dataframe.columns.values[0:-1]
for i in range(len(fit.support_)):
    if fit.support_[i]:
        print(names[i])
```
