```python
from datetime import datetime
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6
```

```python
# Lendo o arquivo e definindo o index
df = pd.read_csv('italy_earthquakes_from_2016-08-24_to_2016-11-30.csv').set_index('Time')

# Convertendo o index de float64 para datetime
df.index = pd.to_datetime(df.index)

# imprimendo com 7 registros
df.head(n=7)
```

```python
# Imprimindo um gráfico com a média de terremotos a cada dia 

# df é o dataframe - magnitude é a coluna - resample vamos regerar uma amostra usando o campo dada - vamos appply(aplicar)
# uma média e plotar um grafico com plot
df['Magnitude'].resample('D').apply([np.mean]).plot()

# Colocando um titulo
plt.title('Magnitude Média Por Dia')

# Colocando um label no eixo y
plt.ylabel('Magnitude')
```

```python
# Imprimindo um gráfico com a média de terremotos a cada 2 dias

# df é o dataframe - magnitude é a coluna - resample vamos regerar uma amostra usando o campo dada - vamos appply(aplicar)
# uma média e plotar um grafico com plot
df['Magnitude'].resample('2D').apply([np.mean]).plot()

# Colocando um titulo
plt.title('Magnitude Média A Cada Dois Dias')

# Colocando um label no eixo y
plt.ylabel('Magnitude')
```

```python
# Filtrando os dados para plotar no grafico

magn3 = df.loc[df['Magnitude'] >= 3.0, 'Magnitude']
```

```python
pd.Series.rolling(magn3, window=25).mean().plot(style='-g')
plt.title('Média Móvel da Magnitude')
plt.ylabel('Magnitude')
plt.xlabel('Tempo')
```
